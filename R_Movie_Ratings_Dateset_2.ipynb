```r
movies <- read.csv('P2-Section6-Movie-Ratings.csv')
```

```r
head(movies)
colnames(movies) <- c('Film', 'Genre', 'CriticsRatings', 'AudienceRatings', 'Budget', 'Year')
colnames(movies) 
```

```r
tail(movies)
str(movies)
movies$Year = factor(movies$Year)
```

```r
library(ggplot2)

q <- ggplot(data = movies, aes(x=CriticsRatings, y=AudienceRatings, colour = Genre))
q + geom_point()
#q + geom_density()
```

```r
# overriding aes
q + geom_point(aes(color = Year))
q + geom_point(aes(x = Budget)) # aes needs category
q + geom_point(size = 3) # geom_point needs attribute
```

```r
# ---------- Histogram and density charts
q <- ggplot(data = movies, aes(x=Budget, fill = Genre))
q + geom_histogram(binwidth = 10, colour = "Black")

q + geom_histogram(aes(x = AudienceRatings), binwidth = 10, colour = "Black") # normal distribution
q + geom_histogram(aes(x = CriticsRatings), binwidth = 10, colour = "Black") # more uniform disribution

q + geom_density()
q + geom_density(position = "stack")
```

```r
# ----------- Statistical transformation 
?geom_smooth()
s <- ggplot(data = movies, aes(x=CriticsRatings, y = AudienceRatings, color = Genre))
s + geom_point() + geom_smooth(fill = NA)
```

```r
# box plots
s <- ggplot(data = movies, aes(x=Genre, y = CriticsRatings, color = Genre))
s <- ggplot(data = movies, aes(x=Genre, y = AudienceRatings, color = Genre))
s + geom_boxplot()
s + geom_boxplot(size = .9)
s + geom_boxplot(size = .9) + geom_point()
s + geom_boxplot(size = .9) + geom_jitter()

s + geom_jitter() + geom_boxplot(size = .9, alpha = .4)  # change order and add transparency
```

```r
# using facets
t <- ggplot(data = movies, aes(x=Budget))
t + geom_histogram(binwidth = 10, aes(fill = Genre), colour = "Black") +
  facet_grid(Genre~.) # we don't want uniform at y-coordinate, next one solves
```

```r
t + geom_histogram(binwidth = 10, aes(fill = Genre), colour = "Black") +
  facet_grid(Genre~., scales = "free") 
```

```r
# scatter plots

w <- ggplot(data = movies, aes(x=CriticsRatings, y = AudienceRatings, color = Genre))
w + geom_point(size = 3) + 
  facet_grid(Genre~.)
```
